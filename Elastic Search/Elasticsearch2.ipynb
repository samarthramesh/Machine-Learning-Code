```python
from elasticsearch import Elasticsearch
from elasticsearch import helpers
```

```python
es = Elasticsearch([{'host': 'localhost','port': 9200}])
```

```python
f = open('/Users/apple/Downloads/recipe1M_layers/layer1.json')
```

```python
import json
data = json.load(f)
#loading data
```

```python
#creating a list in the format needed for bulk indexing
actions = []  
for the_instance in data:
    actions.append({
        "_index": 'food1',
        "_type": 'Recipes',
        "_id": the_instance['id'],
        "_source": the_instance
     })    

  
```

```python
#using helpers to index the file
es = Elasticsearch()
results = helpers.bulk(es, actions, index='food1', doc_type='Recipes', refresh=True)
```
