```python
from elasticsearch import Elasticsearch
from elasticsearch import helpers
import json
```

```python
es = Elasticsearch([{'host': 'localhost','port': 9200}])
```

```python
es
```

```python
file = open('Data/recipes_with_nutritional_info.json')
data = json.load(file)
```

```python
print(data[1])
```

```python
actions = []  
for the_instance in data:
    actions.append({
        "_index": 'food',
        "_type": 'Recipes',
        "_id": the_instance['id'],
        "_source": the_instance
     })    
```

```python
results = helpers.bulk(es, actions, index='food', doc_type='Recipes', refresh=True)
```

```python
res= es.get(index='food',doc_type='Recipes',id="00051d5b9d")
```
